```python
import glob

# convet YOLO format (x_center, y_center, width, height) to normal format (xmax, ymax, xmin, ymin)
def normalize_to_pixel_coordinates(normalized_bbox, image_shape): 
    _, center_x, center_y, width, height = normalized_bbox
    image_width = image_shape[2]
    image_height = image_shape[1]
    x_min = int((center_x - 0.5 * width) * image_width)
    y_min = int((center_y - 0.5 * height) * image_height)
    x_max = int(x_min + (width * image_width))
    y_max = int(y_min + (height * image_height))
    return (x_min, y_min, x_max, y_max)

def extracting_bboxes(image_path, bboxes_path):
    image = read_image(image_path)
    file = open(bboxes_path, 'r').readlines() # the file contains bboxes of YOLO format
    normalized_bboxes = [[int(str_num) if i==0 else float(str_num) for i, str_num in enumerate(line.rstrip().split(" "))] for line in file]
    bboxes = [normalize_to_pixel_coordinates(normalized_bbox, image.shape) for normalized_bbox in normalized_bboxes]
    return torch.tensor(bboxes)

image_paths = sorted(glob.glob("data/solar_pannels/*.jpg"))
bboxes_paths = sorted(glob.glob("data/solar_pannels/*.txt"))

bboxes = extracting_bboxes(image_paths[0], bboxes_paths[0])
bboxes
```
